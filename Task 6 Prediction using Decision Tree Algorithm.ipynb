### Step 1. Importing libraries and Dataset

```python
# Importing the needed libraries
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings as wg
wg.filterwarnings("ignore")
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn import metrics

#importing dataset
Iris=pd.read_csv("Iris.csv")
```

### Step 3. Training the Model

```python
features_vector = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]
x=Iris.loc[:,features_vector].values
x
```

```python
y=Iris.Species
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
Classifier= DecisionTreeClassifier(max_depth=2,random_state=0)
```

```python
Classifier.fit(x_train,y_train)
```

```python
Classifier.predict(x_test[0:3])
```

```python
score = Classifier.score(x_test,y_test)
score
```

```python
#creating Metrics
M = metrics.classification_report(y_test,Classifier.predict(x_test))
print(M)
```

### Step 4. Data Visualization

```python
# let's plot pair plot to visualise the attributes all at once
sns.pairplot(Iris, hue='Species');
```

```python
cm=metrics.confusion_matrix(y_test, Classifier.predict(x_test))
plt.figure(figsize=(8,8))
sns.heatmap(cm, annot=True,
           cmap = "Greens");
plt.ylabel("Actual label", fontsize = 15);
plt.xlabel("Predicted label", fontsize = 15);
plt.title("Accuracy Score: {}".format(score), size= 15);
plt.tick_params(labelsize=15)
```

```python
#list of value to try for max_depth
max_depth_range = list(range(1,6))
accuracy = []
for depth in max_depth_range:
    Classifier= DecisionTreeClassifier(max_depth=depth,random_state=0)
    Classifier.fit(x_train,y_train)
    score = Classifier.score(x_test,y_test)
    accuracy.append(score)
```

```python
#plotting accuracy score depth wise
fig, ax = plt.subplots(nrows =1, ncols=1, figsize= (10,7))
ax.plot(max_depth_range,accuracy,color="k")
ax.set_xlim([1,5])
ax.set_ylim([.50,1.00])
ax.grid(True,axis = "both",zorder =0,linestyle= ":",color='k')
ax.tick_params(labelsize=18)
ax.set_xticks([1,2,3,4,5])
ax.set_xlabel("max_depth", fontsize=20)
ax.set_ylabel("Accuracy", fontsize=20)
fig.tight_layout()
```

```python
# putting the feature names and class names into variables
feature_vector = ["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)"]
class_vector = ["setosa","versicolor","virginica"]
fig, axis = plt.subplots(nrows=1, ncols=1, figsize=(20,10), dpi=300)
tree.plot_tree(Classifier,feature_names=feature_vector,class_names=class_vector,filled= True);
```
